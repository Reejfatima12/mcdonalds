```python
def plot_financial_indicators_boxplot(data_file):
    """
    Create a box plot of financial indicators from a given CSV file.

    Parameters:
    - data_file (str): Path to the CSV file containing the financial data.

    Returns:
    - None
    """
    # Read the dataset
    df = pd.read_csv(data_file)

    # Define the financial indicators
    indicators = ['Revenue ($B)', 
                  'Earnings ($B)', 
                  'P/E ratio',
                  'P/S ratio', 
                  'P/B ratio', 
                  'Operating Margin (%)', 
                  'EPS ($)',
                  'Total debt ($B)', 
                  'Total liabilities ($B)']

    # Create the box plot
    plt.figure(figsize=(12, 8))
    sns.boxplot(data=df[indicators])
    plt.title('Box Plot of Financial Indicators',fontsize=16)
    plt.xticks(rotation=90,fontsize=12)
    plt.grid(True)
    plt.show()

# Example usage:
plot_financial_indicators_boxplot('McDonalds_financial_statements.csv')
```

```python
def plot_performance_metrics_comparison(data_file):
    """
    Plot the performance metrics comparison over the years.

    Parameters:
    - data_file (str): Path to the CSV file containing the financial data.

    Returns:
    - None
    """
    # Read the dataset
    df = pd.read_csv(data_file)

    # Define custom colors
    colors = ['#808080', '#FF0000', '#FFFF00']

    # Plot the performance metrics comparison with custom colors
    df[['Year', 'Revenue ($B)', 'Earnings ($B)', 'Market cap ($B)']].plot(x='Year', kind='bar', figsize=(14, 10), color=colors)
    plt.title('Performance Metrics Comparison',fontsize=14)
    plt.xlabel('Year',fontsize=14)
    plt.ylabel('Value (in $B)',fontsize=14)
    plt.grid(axis='y')
    plt.show()

# Example usage:
plot_performance_metrics_comparison('McDonalds_financial_statements.csv')
```

```python
def plot_shares_outstanding_over_years(data_file):
    """
    Create a scatter plot showing the shares outstanding over years.

    Parameters:
    - data_file (str): Path to the CSV file containing the financial data.

    Returns:
    - None
    """
    # Read the dataset
    df = pd.read_csv(data_file)

    # Create the scatter plot with orange color
    plt.figure(figsize=(10, 8))
    sns.scatterplot(data=df, x='Year', y='Shares Outstanding ($B)', color='orange')
    plt.title('Shares Outstanding Over Years',fontsize=14)
    plt.xlabel('Year',fontsize=14)
    plt.ylabel('Shares Outstanding ($B)',fontsize=14)
    plt.grid(True)
    plt.show()

# Example usage:
plot_shares_outstanding_over_years('McDonalds_financial_statements.csv')
```

```python
def plot_dividend_trend(data_file):
    """
    Plot the trend of dividends adjusted for stock splits over the years.

    Parameters:
    - data_file (str): Path to the CSV file containing the financial data.

    Returns:
    - None
    """
    # Load the dataset
    data = pd.read_csv(data_file)

    # Prepare data
    years = data['Year'].astype(int)  # Convert years to integers
    dividend_adjusted = data['Dividend (stock split adjusted) ($)']

    # Fit a linear model for Dividend (stock split adjusted)
    dividend_adjusted_coefficients = np.polyfit(years, dividend_adjusted, 1)
    dividend_adjusted_polynomial = np.poly1d(dividend_adjusted_coefficients)

    # Generate a line of best fit for Dividend (stock split adjusted)
    dividend_adjusted_trendline = dividend_adjusted_polynomial(years)

    # Flip the years
    years_flipped = years[::-1]
    dividend_adjusted_flipped = dividend_adjusted[::-1]
    dividend_adjusted_trendline_flipped = dividend_adjusted_trendline[::-1]

    # Create the plot for Dividend (stock split adjusted)
    plt.figure(figsize=(10, 8))
    plt.scatter(years_flipped, dividend_adjusted_flipped, color='black', label='Actual Dividend (Stock Split Adjusted) ($)')
    plt.plot(years_flipped, dividend_adjusted_trendline_flipped, color='red', label=f'Linear Fit: y = {dividend_adjusted_coefficients[0]:.2f}x + {dividend_adjusted_coefficients[1]:.2f}')
    plt.title('Dividend (Stock Split Adjusted) of McDonald\'s Over Years',fontsize=16)
    plt.xlabel('Year',fontsize=14)
    plt.ylabel('Dividend (Stock Split Adjusted) ($)',fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
plot_dividend_trend('McDonalds_financial_statements.csv')
```

```python
def calculate_skewness_and_kurtosis(data_file):
    """
    Calculate the skewness and kurtosis of each column in the dataset.

    Parameters:
    - data_file (str): Path to the CSV file containing the dataset.

    Returns:
    - skewness (pd.Series): Series containing the skewness of each column.
    - kurtosis (pd.Series): Series containing the kurtosis of each column.
    """
    # Read the dataset
    df = pd.read_csv(data_file)

    # Calculate skewness
    skewness = df.skew()

    # Calculate kurtosis
    kurtosis = df.kurtosis()

    return skewness, kurtosis

# Example usage:
skewness, kurtosis = calculate_skewness_and_kurtosis('McDonalds_financial_statements.csv')
print("Skewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
def plot_elbow_curve(data_file, features_columns):
    """
    Plots the elbow curve for determining the optimal number of clusters in K-means clustering.

    Parameters:
        data_file (str): The file path of the CSV containing the data.
        features_columns (list): A list of column names representing the features for clustering.

    Returns:
        None
    """
    # Load your data
    df = pd.read_csv(data_file)

    # Select features for clustering
    features = df[features_columns]

    # Standardize the features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Calculate inertia for a range of k values
    inertia = []
    for k in range(1, 11):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(scaled_features)
        inertia.append(kmeans.inertia_)

    # Plot the elbow curve
    plt.figure(figsize=(10, 8))
    plt.plot(range(1, 11), inertia, marker='o')
    plt.title('Elbow Plot for Determining Optimal Clusters')
    plt.xlabel('Number of Clusters', fontsize=14)
    plt.ylabel('Inertia', fontsize=14)
    plt.grid(True)
    plt.show()

# Example usage
plot_elbow_curve('McDonalds_financial_statements.csv', ['Market cap ($B)', 'Revenue ($B)', 'Earnings ($B)'])
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

def visualize_clusters_with_centroids(data_file, features_columns, n_clusters=3):
    """
    Visualizes clusters and centroids in a 2D scatter plot after reducing the data to 2 dimensions using PCA.

    Parameters:
        data_file (str): The file path of the CSV containing the data.
        features_columns (list): A list of column names representing the features for clustering.
        n_clusters (int): Number of clusters to be formed. Defaults to 3.

    Returns:
        None
    """
    # Load your data
    df = pd.read_csv(data_file)

    # Select features for clustering
    features = df[features_columns]

    # Standardize the features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Reduce the data to 2 dimensions using PCA for visualization
    pca = PCA(n_components=2)
    reduced_df = pca.fit_transform(scaled_features)

    # Applying K-means clustering on the reduced data
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(reduced_df)
    labels = kmeans.labels_
    cluster_centers = kmeans.cluster_centers_

    # Calculate marker sizes based on 'Market cap ($B)' (scaled for better visualization)
    marker_sizes = df['Market cap ($B)'] / df['Market cap ($B)'].max() * 100

    # Plotting the 2D scatter plot with the cluster centers
    plt.figure(figsize=(10, 8))
    plt.scatter(reduced_df[:, 0], reduced_df[:, 1], c=labels, s=marker_sizes, cmap='viridis', alpha=0.6)
    centroids = kmeans.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='x', label='Centroids')
    plt.title('Clusters for Market Cap, Revenue and Earnings', fontsize=16)
    plt.xlabel('PCA Feature 1', fontsize=14)
    plt.ylabel('PCA Feature 2', fontsize=14)
    plt.grid(True)
    plt.show()

# Example usage
visualize_clusters_with_centroids('McDonalds_financial_statements.csv', ['Market cap ($B)', 'Revenue ($B)', 'Earnings ($B)'], n_clusters=3)
```
